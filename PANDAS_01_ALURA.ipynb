## Funções unique() e nunique():

As funções `.unique()` e `.nunique()` do pandas são usadas para trabalhar com **valores únicos** em uma série (coluna) de um DataFrame.  

### 🔹 **1. `.unique()` – Retorna os valores únicos**
A função `.unique()` retorna uma **lista de valores únicos** que aparecem em uma coluna, **sem contar quantas vezes eles aparecem**.

#### 🛠 **Exemplo**
```python
import pandas as pd

df = pd.DataFrame({'Tipo': ['A', 'B', 'A', 'C', 'B', 'A', 'D']})

valores_unicos = df['Tipo'].unique()
print(valores_unicos)
```
**Saída:**
```python
['A' 'B' 'C' 'D']
```
✔ Retorna um array NumPy contendo os valores distintos da coluna.  

---

### 🔹 **2. `.nunique()` – Conta quantos valores únicos existem**
A função `.nunique()` retorna **o número total de valores únicos** em uma coluna.

#### 🛠 **Exemplo**
```python
num_unicos = df['Tipo'].nunique()
print(num_unicos)
```
**Saída:**
```python
4
```
✔ Retorna um número inteiro indicando **quantos valores únicos** existem.  

---

### ✅ **Resumo**
| Função         | O que faz? |
|---------------|------------|
| `.unique()`   | Retorna um array com os valores únicos da coluna. |
| `.nunique()`  | Retorna a quantidade de valores únicos na coluna. |

Se quiser apenas **saber quantos valores diferentes existem**, use `.nunique()`.  
Se quiser **ver quais são esses valores**, use `.unique()`.

###Existem diversas formas de tratar dados nulos com o Pandas. Algumas das principais formas são:

## Para saber mais: criando colunas

###Existem diversas maneiras de criar colunas em um DataFrame no Pandas. Alguns exemplos são:

###1 - Atribuição direta de valores a uma nova coluna:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = [7, 8, 9]
df
````

###2 - Criação de uma coluna a partir de operações entre outras colunas:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']
df
````

### 3 - Utilização do método assign() para criar uma nova coluna:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df = df.assign(C=[7, 8, 9])
df
````

### 4 - Utilização do método apply() para aplicar uma função a uma coluna existente e criar uma nova coluna:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'].apply(lambda x: x * 2)
df
````
