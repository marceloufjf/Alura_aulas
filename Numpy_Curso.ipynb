## **Arrays**

###Arrays em Python são estruturas de dados que permitem armazenar múltiplos valores em uma única variável, organizados de forma sequencial. Eles são semelhantes às listas, mas têm algumas diferenças importantes, especialmente em termos de funcionalidade e performance.

##**Características dos Arrays:**

##Homogeneidade:

###Em arrays (diferentemente de listas), todos os elementos devem ser do mesmo tipo. Isso é especialmente útil para processamento numérico.

##Indexação:

###Assim como listas, os arrays são indexados. Você pode acessar elementos individuais usando índices que começam em 0.

##Uso Comum:

###Arrays são comumente usados em contextos onde operações numéricas rápidas e eficientes são necessárias, como em computação científica.

##Definição de Arrays:

###Para criar arrays em Python, é comum usar o módulo **array** da biblioteca padrão ou bibliotecas externas como **NumPy**, que oferecem mais funcionalidade.

##Exemplo com o módulo array:

```python
import array

# Criar um array de inteiros
meu_array = array.array('i', [1, 2, 3, 4, 5])

# Acessar elementos
print(meu_array[0])  # Saída: 1

# Adicionar elementos
meu_array.append(6)
print(meu_array)  # Saída: array('i', [1, 2, 3, 4, 5, 6])

# Remover elementos
meu_array.remove(3)
print(meu_array)  # Saída: array('i', [1, 2, 4, 5, 6])
```

##Exemplo com NumPy:

```python
import numpy as np

# Criar um array
meu_array = np.array([1, 2, 3, 4, 5])

# Acessar elementos
print(meu_array[0])  # Saída: 1

# Operações vetorizadas
print(meu_array * 2)  # Saída: [ 2  4  6  8 10]

# Adicionar outro array
outro_array = np.array([5, 4, 3, 2, 1])
print(meu_array + outro_array)  # Saída: [ 6  6  6  6  6]
```

##Aqui está um exemplo de como transformar uma lista em um array Numpy:

```python
import numpy as np

# cria uma lista
lista = [1, 2, 3, 4, 5]

# transforma a lista em um array Numpy
array = np.array(lista)

print("Lista:", lista)
print("Array:", array)
```

##**Comparativo de performance: listas vs arrays.**

###Focando na eficiência, podemos comparar o tempo levado para efetuar um cálculo utilizando listas e arrays.

```python
import numpy as np
import time

# cria uma lista com 1000000 elementos
lista = list(range(0,101,2))

# transforma a lista em um array Numpy
array = np.array(lista)

# começa a cronometrar o tempo para a operação com a lista
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento da lista
lista_quadrado = [i**2 for i in lista]

# para o cronômetro
list_time = time.time() - start_time

# começa a cronometrar o tempo para a operação com o array
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento do array
array_quadrado = array**2

# para o cronômetro
array_time = time.time() - start_time

print("Lista:", lista)
print("\nArray:", array)
print("\nLista ao quadrado:", lista_quadrado)
print("\nArray ao quadrado:", array_quadrado)
print("\nTempo da operação com a lista: ", list_time)
print("\nTempo da operação com o array: ", array_time)

# Obs: quando as operações são feitas com lista ou arrays menores, a diferença de tempo não é relevante.
```

##Para saber mais: cópias de arrays

###Ao trabalharmos com arrays ou outras variáveis em Python, eventualmente fazemos cópias das variáveis que estamos lidando. Quando fazemos a cópia de uma variável, temos que ter em mente o que acontece com ela para evitar que usemos determinados valores pensando que são outros valores. Para lidar com cópias dos arrays, a biblioteca Numpy possui uma função específica.

###Neste exemplo, possuo um array com o nome preco_imoveis que desejo fazer uma cópia.

###O objetivo é colocar os valores deste array em um de nome específico para o preço de imóveis em São Paulo. Vamos fazer a cópia usando o sinal de igual (=).

```python
preco_imoveis = np.array([10000,120000,11000,200000])

preco_imoveis_sao_paulo = preco_imoveis

print(preco_imoveis)
print(preco_imoveis_sao_paulo)
```

###Se você verificar os valores na nova variável preco_imoveis_sao_paulo, verá que eles são os mesmos da variável preco_imoveis. Mas, já que "guardamos" os preços na variável preco_imoveis_sao_paulo, que tal substituir o valor do primeiro imóvel do array preco_imoveis com um novo que acabou de entrar para o nosso cadastro?

###Se você tentar reproduzir o processo feito até agora, verá que essa alteração no array preco_imoveis também alterou a primeira posição do array preco_imoveis_sao_paulo para o valor de 120000.



```python
preco_imoveis[0] = 120000

print(preco_imoveis)
print(preco_imoveis_sao_paulo)
```

###Para criar um array desvinculado do array original existe uma função da numpy, a np.copy(). Para criar a cópia utilizando ela bastaria seguir esse código:

```python
preco_imoveis_2 = np.array([10000,120000,11000,200000])
preco_imoveis_sao_paulo_2 = np.copy(preco_imoveis_2)

preco_imoveis_2[0] = 50000


print(preco_imoveis_2)   # Repare que desta vez só foi alterado uma variável.
print(preco_imoveis_sao_paulo_2) # Esse artifício é útil quando se manipula cadastros.
```

###Assim, mesmo que alterarmos o array preco_imoveis, isso não afetará o array preco_imoveis_sao_paulo.

###Isso ocorre porque, quando usamos o sinal de igual para copiar o array, temos um novo array que aponta para os mesmos locais na memória. Já quando usamos o np.copy(), os locais na memória são diferentes.

## - Multiplicação

###Agora iremos calcular o coeficiente angular de forma mais assertiva e direta:

a = Coeficiente angular

n = Número de elementos

Y = Moscow

X = Datas

O coeficiente angular pode ser obtido usando a equação:



â =  n*Soma(Xi * Yi) - Soma(Xi) * Soma(Yi)
_____________________________________________
         n*Soma(Xi^2) - (Soma(Xi))^2




```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
X**2 # Foi o usado este artifício "**", pois com "np.power()" a fórmula ficaria comprida.
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X))**2)
print(f'O coeficiente angular é: a = {a:.2f}')
```

### Agora vamos calular o coeficiente linear "b" da equação Y = ax + b

Fórmula:

b = Media(Yi) - a*Media(Xi)

```python
b = np.mean(Y) - a*np.mean(X)
print(f'O coeficiente linear é: b = {b:.2f}')
```
